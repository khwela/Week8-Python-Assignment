```python
# Load the dataset
us_covid19_daily = pd.read_csv(r'C:\2025 COHORT FEB\PYTHON\Python\COVID-19\us_covid19_daily.csv')

# Display the first few rows of the dataset
print(us_covid19_daily.head())

# Plot a line graph for positive cases over time
plt.figure(figsize=(10, 6))
plt.plot(us_covid19_daily['date'], us_covid19_daily['positive'], label='Positive Cases', color='blue')
plt.xlabel('Date')
plt.ylabel('Positive Cases')
plt.title('Positive COVID-19 Cases Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.show()

# Plot a histogram for the distribution of deaths
plt.figure(figsize=(10, 6))
plt.hist(us_covid19_daily['death'].dropna(), bins=30, color='red', alpha=0.7)
plt.xlabel('Deaths')
plt.ylabel('Frequency')
plt.title('Distribution of COVID-19 Deaths')
plt.show()

# Plot a bar chart for hospitalized cases
plt.figure(figsize=(10, 6))
sns.barplot(x='date', y='hospitalizedCurrently', data=us_covid19_daily, color='green')
plt.xlabel('Date')
plt.ylabel('Hospitalized Currently')
plt.title('Hospitalized Cases Over Time')
plt.xticks(rotation=45)
plt.show()
```
